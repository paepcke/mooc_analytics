```python
tst_data = pn.DataFrame(np.asarray([
            ['vid1','s1',1,1],
            ['vid1','s1',2,1],
            ['vid1','s1',3,1],
            ['vid2','s1',1,1],
            ['vid2','s1',2,1],
            ['vid2','s1',3,1],
            ['vid1','s2',1,1],
            ['vid1','s2',2,1],
            ['vid1','s2',3,1],
            ['vid2','s2',1,1],
            ['vid2','s2',2,1]
            ]), columns=['videoId','anon_screen_name','second','numViews']
)

            
            
```

```python
grpByObj = pn.to_numeric(tst_data['numViews']).groupby([tst_data['videoId'],tst_data['anon_screen_name']])
```

```python
grpByObj.sum()
```

```python
# Series sums of views by student by video:
sums_stud_vid = grpByObj.sum()
mean_stud_vid = sums_stud_vid.mean()
std_stud_vid  = sums_stud_vid.std()
print('Mean number of views across all videos: %s' % sums_stud_vid.mean())
print('Stdev of number of views across all videos: %s' % sums_stud_vid.std())
low_watcher_mask   = sums_stud_vid < mean_stud_vid - 2*std_stud_vid
print('Low-watchers: %s' % low_watcher_mask)
norm_watcher_mask  = (sums_stud_vid > mean_stud_vid - 2*std_stud_vid) & (sums_stud_vid < mean_stud_vid + 2*std_stud_vid)
print('Normal-watchers: %s' % norm_watcher_mask)
high_watcher_mask = sums_stud_vid > mean_stud_vid + 2*std_stud_vid
print('High-watchers: %s' % high_watcher_mask)
```

```python
learners = tst_data['anon_screen_name'].unique()
learners
```

```python
# Prepare a table that records majority viewing habits
# for each learner:
watch_habits = pn.DataFrame(learners, columns=['anon_screen_name'])
watch_habits
```

```python
# Add column 'watch_level' to df watch_habits.
# Result will be:
#   anon_screen_name    watch_level
#         s0                norm
#         s1                 low
#         s2                hight
#
# Watch level is low for a learner if:
#    for the majority of videos V the learner watched
#    fewer seconds than mean-2*SD over all learners who
#    watched V.
# Watch level is normal for a learner if: 
#    for the majority of videos V the learner watched
#    between mean-2*SD<mean<mean+2SD seconds.
# Watch level is high for a learner if:
#    for the majority of videos V the learner watched
#    >mean+2SD seconds.

vid_watch_levels = []
for watcher_id in watch_habits['anon_screen_name']:
    # Sum over the masks thats indicate low/norm/hi levels
    # of vid views:
    watch_levels = pn.Series([high_watcher_mask[:,watcher_id].sum(),
                              norm_watcher_mask[:,watcher_id].sum(),
                              low_watcher_mask[:,watcher_id].sum()
                             ])
    # Pick the proper indicator based on which of the 
    # sums above is highest:
    vid_watch_levels.append(['low', 'norm', 'high'][watch_levels.idxmax()])

vid_watch_levels = pn.Series(vid_watch_levels)
# Add the 'watch_level' column to the watch_habits
# table, which only has column 'anon_screen_name':
watch_habits = watch_habits.assign(watch_level=vid_watch_levels)
watch_habits
```
